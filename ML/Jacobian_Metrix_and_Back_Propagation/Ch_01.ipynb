```python
# Learning rate and Gradient-base Learning
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3, 3, 100)
function_y = function_x**2

fig, ax= plt.subplots(figsize=(10, 5))
ax.plot(function_x, function_y)

x = 2
lr = 0.1

for _ in range(10):
    dy_dx = 2*x
    x = x - lr*dy_dx
    y = x**2

    ax.scatter(x, y, color='red', s = 100)
```

```python
# Effectiveness of Gradient
# Increases from a small differential coefficents to a large differential coefficient
import numpy as np
import matplotlib.pyplot as plt

function_x = np.linspace(-3, 3, 100)
function_y = 2*function_x**2

fig, ax= plt.subplots(figsize=(10, 5))
ax.plot(function_x, function_y)

x = 1
lr = 1

for _ in range(10):
    dy_dx = 4*x
    x = x - lr*dy_dx
    y = 2*x**2

    ax.scatter(x, y, color='red', s = 100)
```
