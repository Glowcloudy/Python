plot 종류별 사용법

- bar plot

```python
# bar plot
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6])
plt.show()
```

```python
# width 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width=0.5)
plt.show()
```

```python
# bottom 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom=2)
plt.show()
```

```python
# align 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align='edge')
plt.show()
```

histogram plot

- data의 분포를 나타낼 때 많이 사용

```python
# histogram
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500))
plt.show()
```

```python
# bins
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
# ax.hist(np.random.randn(500), [-3, -2, -1, 0, 1, 2, 3])
ax.hist(np.random.randn(500), np.linspace(-2, 2, num=40))
plt.show()
```

```python
# range
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), range(-2, 2))
plt.show()
```

```python
# density
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=40, range=(-2, 2), density=True)
plt.show()
```

scatter plot
- data간 유사도와 상관관계를 파악할 때 많이 사용

```python
import pandas as pd
```

```python
iris_df = pd.read_csv('./Iris.csv')
iris_df.head()
```

```python
fig = plt.figure(figsize=(7, 7))

ax = fig.add_subplot()
ax.scatter(x = iris_df['PetalLengthCm'], y = iris_df['PetalWidthCm'])
plt.show()
```

한 subplot에 여러개 graph 그리기

```python
# 3개의 graph 동시에 그리기
fig = plt.figure()

ax = fig.add_subplot()
ax.plot([1, 1.5, 2])
ax.plot([1, 2, 3])
ax.plot([1, 2.5, 4])

plt.show()
```

```python
# line plot과 bar plot 동시에 그리기
fig = plt.figure()

ax = fig.add_subplot()
ax.plot([1, 2, 3], [1, 2, 3])
ax.bar([1, 2, 3], [1, 2, 3])
plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))

ax = fig.add_subplot()

for species in iris_df['Species'].unique():
    iris_sub = iris_df[iris_df['Species']==species]
    ax.scatter(x=iris_sub['PetalLengthCm'], y=iris_sub['PetalWidthCm'])

plt.show()
```

여러개 subplot 한번에 만들기 

```python
# plt.Subplot 활용
fig, axes = plt.subplots(2, 3, figsize=(10, 5))
axes[0, 1].plot([1, 2, 3])
axes[1, 2].plot([3, 2, 1])
plt.show()
```

```python
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10, 5))
axes[0, 0].hist(np.random.randn(200), bins=20)
axes[0, 1].bar(np.arange(10), np.random.randn(10)*10+10)
axes[1, 0].scatter(np.arange(30), np.arange(30)+3 *np.random.randn(30))
axes[1, 1].plot(np.random.randn(50).cumsum()*5+20)
plt.show()
```

```python
# y축 눈금 공유
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10, 5), sharey=True)
axes[0, 0].hist(np.random.randn(200), bins=20)
axes[0, 1].bar(np.arange(10), np.random.randn(10)*10+10)
axes[1, 0].scatter(np.arange(30), np.arange(30)+3 *np.random.randn(30))
axes[1, 1].plot(np.random.randn(50).cumsum()*5+20)
plt.show()
```

```python
# subplot 간격 조정
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10, 5), sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50)
plt.subplots_adjust(wspace=0, hspace=0)
plt.show()
```

색상 지정하기

```python
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10, 5), sharey=True)
axes[0, 0].hist(np.random.randn(200), bins=20, color='r')
axes[0, 1].bar(np.arange(10), np.random.randn(10)*10+10, color='seagreen')
axes[1, 0].scatter(np.arange(30), np.arange(30)+3, np.random.randn(30), color='#0000ff')
axes[1, 1].plot(np.random.randn(50).cumsum()*5+20, color='brown')
plt.show()
```

label 지정하기

```python
# 범례 표시
fig = plt.figure(figsize=(7, 7))

ax = fig.add_subplot()

for species in iris_df['Species'].unique():
    iris_sub = iris_df[iris_df['Species']==species]
    ax.scatter(x=iris_sub['PetalLengthCm'], y=iris_sub['PetalWidthCm'], label=species)

plt.legend()
plt.show()
```

제목, 축이름, 눈금 설정하기

```python
fig = plt.figure(figsize=(7, 7))

ax = fig.add_subplot()

for species in iris_df['Species'].unique():
    iris_sub = iris_df[iris_df['Species']==species]
    ax.scatter(x=iris_sub['PetalLengthCm'], y=iris_sub['PetalWidthCm'], label=species)

ax.set_title('Iris')
xticks = np.arange(1, 7.5, 0.5)
ax.set_xticks(xticks)
ax.set_xlabel('PetalLengthCm')
ax.set_ylabel('PetalWidthCm')

plt.legend()
plt.show()
```
